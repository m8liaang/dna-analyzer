```python
# Make sure DNA input is valid

def valid(sequence):

  allowed_characters = {'A', 'G', 'C', 'T'}

  sequence = sequence.upper().replace(" ", "")

  for char in sequence:
    if char not in allowed_characters:
      return False

  return sequence
```

```python
# DNA Base Counter

def base_counter(sequence):
  A_count, G_count, C_count, T_count = 0, 0, 0, 0

  for letter in sequence:
    if letter == 'A':
      A_count += 1
    elif letter =='G':
      G_count += 1
    elif letter == 'C':
      C_count += 1
    else:
      T_count += 1

  return print(f"\nBase Counts:\nA: {A_count}, G: {G_count}, C: {C_count}, T: {T_count}")
```

```python
# GC Content Calculator

def GC_content(sequence):
  G_counts = sequence.count('G')
  C_counts = sequence.count('C')

  GC_content = ((G_counts + C_counts) / len(sequence)) * 100

  return print(f"\nGC Content: {round(GC_content, 2)}%")
```

```python
# Complementary Strand Generator

def DNA_complement(sequence):
  complements = {
      'A' : 'T',
      'G' : 'C',
      'C' : 'G',
      'T' : 'A'
  }
  allowed_characters = {'A', 'G', 'C', 'T'}
  DNA_complements = ''

  for char in sequence:
    for key, value in complements.items():
      if key == char:
        DNA_complements += str(complements[f'{key}'])

  return print(f'\nComplementary Strand: {DNA_complements}')
```

```python
# mRNA Sequence Generator

def RNA_generator(sequence):
  allowed_characters = {'A', 'G', 'C', 'T'}

  for char in sequence:
    new_sequence = sequence.replace('T', 'U')

  return print(f'\nmRNA strand: {new_sequence}')
```

```python
# Main Function

def main():

  DNA_string = input("Enter a DNA sequence: ")

  sequence = valid(DNA_string)

  if sequence is False:
    print("DNA sequence is not valid.")
  else:
    base_counter(sequence)
    GC_content(sequence)
    DNA_complement(sequence)
    RNA_generator(sequence)

main()
```
